```python
import os
import torch

# Create a JAX folder in Drive if it doesn't exist
jax_folder = "/content/drive/MyDrive/JAX_HOT"
os.makedirs(jax_folder, exist_ok=True)

# Outfit variations
outfits = [
    "wearing only a white towel around waist",
        "in tight red jockstrap with visible bulge",
            "posing shirtless in low-rise jeans",
                "in black leather harness and briefs",
                    "in gym shorts with sweat glistening on chest",
                        "in tight white briefs standing by window light",
                            "in bed with open button-down shirt",
                                "in low-cut tank top and short shorts",
                                    "in tight speedo at the beach",
                                        "in bathrobe slipping off shoulders"
                                        ]

base_prompt = "handsome muscular young man, short styled brown hair, light stubble, piercing blue eyes, smooth skin, defined abs, photorealistic, realistic lighting, 8k, RealVisXL V5.0, SDXL"

# Generate 10 JAX images with outfit variations
for i, outfit in enumerate(outfits):
    prompt = f"{base_prompt}, {outfit}"
    seed = 7095590365040601 + i  # Slightly vary the seed
    generator = torch.manual_seed(seed)
    image = pipe(prompt, generator=generator).images[0]
    image.save(f"{jax_folder}/jax-hot-{i+1}.png")
```
